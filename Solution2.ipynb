```python
import pandas as pd
```

```python
data1 = pd.read_csv("/content/data1.csv")
data2 = pd.read_csv("/content/data2.csv")
data2 = data2.drop(columns=['name'])
data3 = pd.read_csv("/content/data3 (1).csv")
data2.head()
```

```python
train_data = pd.merge(data1, data2, on='id')
train_data = pd.merge(train_data, data3, on='id')
```

```python
train_data = train_data.drop(columns=['username', 'mail', 'address', 'birthdate','sex'])
train_data.to_csv('modified_train_data.csv', index=False)
```

```python
train_data.head()
```

```python
X_train = train_data.drop(columns=['Well Being','name','id'])
y_train = train_data['Well Being']
```

```python
X_train.fillna(X_train.mode(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_val)
accuracy = accuracy_score(y_val, y_pred)
print(f'Validation Accuracy: {accuracy}')
```
